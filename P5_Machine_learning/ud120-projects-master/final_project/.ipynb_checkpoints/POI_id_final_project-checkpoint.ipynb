### Creating new features

```python
# Creating new features
extended_features_list = []
extended_features_list = [i for i in features_list]

# The if/else statements in the for loop are included to create "NaN" values if any of the inputs are "NaN"
# This is to conform to the feature_format.py function
for pax in data_dict:
    
    if (data_dict[pax]['from_this_person_to_poi'] != "NaN") and (data_dict[pax]['to_messages'] != "NaN"):
        data_dict[pax]['proportion_to_msg_poi'] = round(float(data_dict[pax]['from_this_person_to_poi']) / \
                                                (float(data_dict[pax]['to_messages'])), 3)
    else:
        data_dict[pax]['proportion_to_msg_poi'] = "NaN"
    
    if (data_dict[pax]['from_poi_to_this_person'] != "NaN") and (data_dict[pax]['from_messages'] != "NaN"):
        data_dict[pax]['proportion_from_msg_poi'] = round(float(data_dict[pax]['from_poi_to_this_person']) / \
                                                float(data_dict[pax]['from_messages']), 3)
    else:
        data_dict[pax]['proportion_from_msg_poi'] = "NaN"
    
    if (data_dict[pax]['bonus'] != "NaN") and (data_dict[pax]['salary'] != "NaN"):
        data_dict[pax]['bonus_to_salary_ratio'] = round(float(data_dict[pax]['bonus']) / \
                                                float(data_dict[pax]['salary']), 3)
    else:
        data_dict[pax]['bonus_to_salary_ratio'] = "NaN"
    
    if (data_dict[pax]['expenses'] != "NaN") and (data_dict[pax]['salary'] != "NaN"):
        data_dict[pax]['expenses_to_salary_ratio'] = round(float(data_dict[pax]['expenses'])/ \
                                                float(data_dict[pax]['salary']), 3)
    else:
        data_dict[pax]['expenses_to_salary_ratio'] = "NaN"
    
    if (data_dict[pax]['exercised_stock_options'] != "NaN") and (data_dict[pax]['total_stock_value'] != "NaN"):
        data_dict[pax]['exercised_to_total_stock_ratio'] = round(float(data_dict[pax]['exercised_stock_options'])/ \
                                                float(data_dict[pax]['total_stock_value']), 3)
    else:
        data_dict[pax]['exercised_to_total_stock_ratio'] = "NaN"
    

extended_features_list.append('proportion_to_msg_poi')
extended_features_list.append('proportion_from_msg_poi')
extended_features_list.append('bonus_to_salary_ratio')
extended_features_list.append('expenses_to_salary_ratio')
extended_features_list.append('exercised_to_total_stock_ratio')
```
